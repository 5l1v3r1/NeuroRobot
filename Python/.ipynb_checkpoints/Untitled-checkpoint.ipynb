```python
import numpy as np
import cv2
import time
import requests
import matplotlib.pyplot as plt
```

```python

cap = cv2.VideoCapture("rtsp://admin:admin@192.168.100.1/cam1/h264")

for i in range(10):
    ret, frame = cap.read()
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    plt.imshow(frame)
    plt.show()
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
