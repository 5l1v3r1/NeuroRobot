```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
s0 = np.load("static-0m.npy")
s1 = np.load("static-1m.npy")
s2 = np.load("static-2m.npy")
s3 = np.load("static-3m.npy")
s4 = np.load("static-4m.npy")
i0 = np.load("imshow-0m.npy")
i1 = np.load("imshow-cv2-0m.npy")


plt.plot(i1)
plt.show()
```

```python
plt.figure(figsize=(12,12))
plt.scatter(np.linspace(0,len(i0)-1,len(i0)-1), np.ediff1d(i0), s=1, label="PyAV")
plt.scatter(np.linspace(0,len(i0)-1,len(i1)-1), np.ediff1d(i1), s=1, label="OpenCV" )
plt.ylim(0,0.5)
plt.ylabel("inter-frame delay (seconds)")
plt.xlabel("frames")
plt.title("Comparison of inter-frame delay between OpenCV and PyAV")
plt.legend()
plt.show()
```

```python
x = np.round(np.ediff1d(s0), 3)
y = np.round(np.ediff1d(s1), 3)
```

```python
plt.figure(figsize=(12,12))
plt.hist(x, bins=100, alpha=0.5)  # arguments are passed to np.histogram
plt.hist(y, bins=100, alpha=0.5)  # arguments are passed to np.histogram
plt.xlabel("inter-frame delay (seconds)")
plt.ylabel("frames")
plt.show()
```
